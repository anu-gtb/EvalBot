```python
## Functions for text cleaning techniques
def remove_special_characters(text):
  translator=str.maketrans('','',string.punctuation)
  return text.translate(translator)

def lower_case(text):
  text=text.lower()
  return text

def remove_stopwords(text):
  words=text.split()
  stop_words=set(stopwords.words('english'))
  words=[word for word in words if word not in stop_words]
  return ' '.join(words)
```

```python
## Apply text cleaning functions to training data
train['sentence1']=train['sentence1'].apply(remove_special_characters)
train['sentence1']=train['sentence1'].apply(lower_case)
train['sentence1']=train['sentence1'].apply(remove_stopwords)

train['sentence2']=train['sentence2'].apply(remove_special_characters)
train['sentence2']=train['sentence2'].apply(lower_case)
train['sentence2']=train['sentence2'].apply(remove_stopwords)
```

```python
train.head()
```

```python
## Apply text cleaning functions to validation data
val['sentence1']=val['sentence1'].apply(remove_special_characters)
val['sentence1']=val['sentence1'].apply(lower_case)
val['sentence1']=val['sentence1'].apply(remove_stopwords)

val['sentence2']=val['sentence2'].apply(remove_special_characters)
val['sentence2']=val['sentence2'].apply(lower_case)
val['sentence2']=val['sentence2'].apply(remove_stopwords)
```

```python
val.head()
```

```python
## Reset index of datasets
train=train.reset_index(drop=True)
val=val.reset_index(drop=True)
```
